### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento



```python
loja1_fat = loja1["Preço"].sum()
loja2_fat = loja2["Preço"].sum()
loja3_fat = loja3["Preço"].sum()
loja4_fat = loja4["Preço"].sum()
```

# 2. Vendas por Categoria

```python
loja1_cat = loja1['Categoria do Produto'].value_counts()
loja2_cat = loja2['Categoria do Produto'].value_counts()
loja3_cat = loja3['Categoria do Produto'].value_counts()
loja4_cat = loja4['Categoria do Produto'].value_counts()
```

# 3. Média de Avaliação das Lojas


```python
loja1_med_aval = loja1['Avaliação da compra'].mean()
loja2_med_aval = loja2['Avaliação da compra'].mean()
loja3_med_aval = loja3['Avaliação da compra'].mean()
loja4_med_aval = loja4['Avaliação da compra'].mean()
```

# 4. Produtos Mais e Menos Vendidos

```python
loja1_head_vendas = loja1['Produto'].value_counts().head(5)
loja2_head_vendas = loja2['Produto'].value_counts().head(5)
loja3_head_vendas = loja3['Produto'].value_counts().head(5)
loja4_head_vendas = loja4['Produto'].value_counts().head(5)


loja1_tail_vendas = loja1['Produto'].value_counts().tail(5)
loja2_tail_vendas = loja2['Produto'].value_counts().tail(5)
loja3_tail_vendas = loja3['Produto'].value_counts().tail(5)
loja4_tail_vendas = loja4['Produto'].value_counts().tail(5)
```

```python
print(loja1_head_vendas)
print(loja2_head_vendas)
print(loja3_head_vendas)
print(loja4_head_vendas)
```

# 5. Frete Médio por Loja



```python
loja1_med_frete = loja1['Frete'].mean()
loja2_med_frete = loja2['Frete'].mean()
loja3_med_frete = loja3['Frete'].mean()
loja4_med_frete = loja4['Frete'].mean()
```

```python
print(loja1_med_frete)
print(loja2_med_frete)
print(loja3_med_frete)
print(loja4_med_frete)
```

## Visualizar Vendas por Categoria

### Subtask:
Gerar um gráfico de barras agrupadas para comparar a distribuição de vendas por categoria de produto em cada loja, identificando quais lojas têm maior ou menor concentração em certas categorias.

```python
import matplotlib.pyplot as plt

faturamento_total = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [loja1_fat, loja2_fat, loja3_fat, loja4_fat]
})

faturamento_total.plot(kind='bar', y='Faturamento', x='Loja', title='Faturamento Total por Loja', xlabel='Loja', ylabel='Faturamento', rot=0)
plt.show()
```

## Gerar Gráfico de Frete Médio

### Subtask:
Criar um DataFrame com o frete médio de cada loja e gerar um gráfico de linhas para comparar os custos de frete entre as lojas.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Combine category counts from all stores into a single DataFrame
all_categories = pd.DataFrame({
    'Loja 1': loja1_cat,
    'Loja 2': loja2_cat,
    'Loja 3': loja3_cat,
    'Loja 4': loja4_cat
}).fillna(0)

# Plotting the grouped bar chart
all_categories.plot(kind='bar', figsize=(12, 7))
plt.title('Vendas por Categoria do Produto por Loja')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Loja')
plt.tight_layout()
plt.show()
```

```python
frete_medio_loja = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': [loja1_med_frete, loja2_med_frete, loja3_med_frete, loja4_med_frete]
})

frete_medio_loja.plot(kind='line', x='Loja', y='Frete Médio', title='Frete Médio por Loja', xlabel='Loja', ylabel='Frete Médio')
plt.show()
```
